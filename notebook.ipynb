````python
def mmin(x):
    valeurmin = x[0]
    ind = 0
    for indice, valeur in enumerate(x[1:]):
        if valeur < valeurmin:
            valeurmin = valeur
            ind = indice+1

    return ind, valeurmin

x = [-11, 6, -10, 55]
mmin(x)
````

````python
def mysort(x, type="asc"):
    x = np.asarray(x)
    if type == "asc":        
        fun = mmin
    elif type == "desc":
        fun = mmax
    else: 
        print("error, type should be asc or desc")

    indout = list()
    listindice = np.asarray(range(len(x)))
    
    while True:
        indice_xpl = list()
        for l in listindice:
            if l not in indout:
                indice_xpl.append(l)
    
        i,v = fun(x[indice_xpl])
        ii = listindice[indice_xpl]
        
        indout.append(int(ii[i]))

        if len(indout)>=len(x):
            return indout, np.asarray(x)[indout]
        
x = [-1, 66, -10, 55]
indout, out = mysort(x)
````
