````python
import os

PATH = r'.venv'

Nfiles = 0
Ndirs = 0

for root, dirnames, filenames in os.walk(PATH):
    Ndirs += len(dirnames)
    Nfiles += len(filenames) 

print("Ndirs", Ndirs)
print("Nfiles", Nfiles)
````

````python
from os import walk as oswalk

PATH = r'.venv'

Nfiles = 0
Ndirs = 0

for root, dirnames, filenames in oswalk(PATH):
    Ndirs += len(dirnames)
    Nfiles += len(filenames) 

print("Ndirs", Ndirs)
print("Nfiles", Nfiles)
````
